```python
import pandas as pd
import json
import numpy as np
#import codecs
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import requests
import os
```

```python
df = pd.read_json('yelp/yelp_academic_dataset_business.json', lines=True, encoding = 'utf8')
#with open ("yelp/yelp_academic_dataset_business.json") as f:
 #   data = json.load(f)
#print(data)
df
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
url = 'https://www.nrcs.usda.gov/wps/portal/nrcs/detail/?cid=nrcs143_013696'
browser.visit(url)
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
tablerows = soup.find("table", class_="data").find("tbody").find_all("tr")
staterow = []
allstates = []
for row in tablerows:
    try:
        rowdata = row.find_all("td")
        for td in rowdata:
            staterow.append(td.text.strip())
        #if(len(staterow)>0):
            #state.append(staterow[0])
            #statecode.append(staterow[1])
            #fipscode.append(staterow[2])
        #staterow = []
        allstates.append(staterow)
        staterow = []
    except AttributeError as e:
        print(e)
#print(state)
#print(statecode)
#print(fipscode)
print(allstates)
```

```python
usst = []
for i in range(1, len(allstates)-1):
    currentrow = allstates[i]
    currentstate = currentrow[1]
    usst.append(currentstate)
    
usst
```

```python
us_yelp_data = df.loc[df["state"].isin(usst), :]
dflen = len(us_yelp_data)
#dflen
```

```python
us_yelp_data["categories"] = us_yelp_data['categories'].str.replace(',', ':')
us_yelp_data["address"] = us_yelp_data['address'].str.replace(',', ':')
us_yelp_data["name"] = us_yelp_data['name'].str.replace(',', ':')
us_yelp_data.head(50)

```

```python
#us_yelp_data.drop(columns = ["attributes", "hours"], inplace=True)
us_yelp_data.dropna()
us_yelp_data
```
